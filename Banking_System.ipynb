## Banking System Analogy
- **Class:** Type of bank account,Savings/Current account 
- **Object:** Specific account type, Savings account
- **Attribute:** Account details(account number ,account name) 
- **Methods:** Transaction type(deposit, redrawal,Tansfer, Payment(Bill Payment, E-Banking ,Card Transaction,Wire    Transfers))


## Basic implementation of deposit and withdrawal.

```python
import random
```

```python
class Bank_account:
    def __init__ (self,account_name, account_number, account_balance):
        self.account_name = account_name
        self.account_number = account_number
        self.account_balance =account_balance
    def deposit(self, amount):
        if amount >0:
            self.account_balance += amount 
            print ("the total amount deposited")
    def withdrawal (self,amount):
        if amount >self. account_balance:
            print("inssufficent balance")
        elif amount >50:
                 print("withdrawal successful")
        elif amount > self.account_balance:
             print("withdrawal unsuccessful")
        else:
            self.account_balance -= amount
            print("new balance")
obj = Bank_account('Mimi',random.randint(1,12), 500)
obj.deposit(20)
obj.withdrawal(100)

```

## Introduced minimum balancen checks & improved error handling

```python
class Bank_account:
    def __init__ (self,account_name, account_number, account_balance):
        self.account_name = account_name
        self.account_number = account_number
        self.account_balance =account_balance
    def deposit(self, amount):
        if amount >0:
            self.account_balance += amount 
            print ("deposit successful")
            print ("new balance:", self.account_balance)
        else:
             print("invalid deposit amount")
    def withdrawal (self,amount):
        if amount <= 0:
            print("invalid amount")
        elif amount > self.account_balance:
             print("inssufficient fund for withdrawal ")
        elif self.account_balance - amount < 50:
             print("cannot go below minimum balance of 50")
        else:
            self.account_balance -= amount
            print("withdrawal successful")
            print("new balance:", self.account_balance)
obj = Bank_account('Mimi',random.randint(1,12), 500)
obj.deposit(20)
obj.withdrawal(100)
obj.withdrawal(500)

```
